```python
# 1. Estruturando os Dados ---

# Agrupamos todos os DataFrames das lojas em um único dicionário.
lojas_df = {
    "Loja 1": loja1,
    "Loja 2": loja2,
    "Loja 3": loja3,
    "Loja 4": loja4
}

# Convertendo os DataFrames para listas de dicionários
vendas_lojas = {
    nome_loja: df.to_dict(orient='records')
    for nome_loja, df in lojas_df.items()
}

# 2. Função para Cálculo do Faturamento
def obter_faturamento(vendas: list) -> float:
    """
    Soma o valor total das vendas de uma loja.
    """
    soma_total = 0.0
    for item in vendas:
        soma_total += item['Preço']
    return soma_total

# 3. Cálculo e Exibição dos Resultados ---

print("=== Faturamento Total por Loja ===")

# Armazena os resultados para uso posterior (gráficos, por exemplo)
faturamentos = {}

for nome_loja, lista_vendas in vendas_lojas.items():
    total = obter_faturamento(lista_vendas)
    faturamentos[nome_loja] = total
    print(f"{nome_loja}: R$ {total:,.2f}")
```

# 2. Vendas por Categoria

```python
# 2. Vendas por Categoria

# 1. Contagem
def contar_categorias(lista_vendas: list) -> dict:
    """
    Conta a ocorrência de cada categoria de produto em uma lista de vendas.
    """
    contagem = {}

    for venda in lista_vendas:
        categoria = venda['Categoria do Produto']
        contagem[categoria] = contagem.get(categoria, 0) + 1

    return contagem


# 2. Resultados por loja
print("--- Contagem de Vendas por Categoria ---")

# Dicionário para guardar todos os resultados
resultados_categorias = {}

# Loop para calcular e imprimir os resultados
for nome_loja, lista_vendas in vendas_lojas.items():
    # 1. Calcula a contagem para a loja atual
    contagem_loja = contar_categorias(lista_vendas)

    # 2. Armazena o resultado
    resultados_categorias[nome_loja] = contagem_loja

    # 3. Exibe de forma formatada
    print(f"\n{nome_loja.upper()}:")
    print(contagem_loja)

# Agora 'resultados_categorias' contém os dados prontos para gráficos
```

```python
# 1. Criar nossa função de contagem (reutilizável!)
def contar_categorias(lista_vendas: list) -> dict:
    """
    Conta a ocorrência de cada categoria de produto em uma lista de vendas.
    """
    contagem = {}

    for venda in lista_vendas:
        categoria = venda['Categoria do Produto']
        contagem[categoria] = contagem.get(categoria, 0) + 1

    return contagem


# 2. Criar uma função "helper" para ordenar nossos dicionários
def ordenar_contagem(dicionario_contagem: dict) -> list:
    """
    Ordena um dicionário de contagem pelo valor (contagem),
    do maior para o menor.
    Retorna uma lista de tuplas (chave, valor).
    """
    itens_ordenados = sorted(
        dicionario_contagem.items(),
        key=lambda item: item[1],  # Ordena pelo valor da contagem
        reverse=True
    )
    return itens_ordenados


# 3. Calcular e Ordenar os resultados (de forma dinâmica)
print("--- Vendas por Categoria (Mais Populares) ---")

# Dicionário para guardar todos os resultados finais e ordenados
resultados_categorias_ordenados = {}

# Loop único para calcular, ordenar e imprimir os resultados
for nome_loja, lista_vendas in vendas_lojas.items():

    # 1. Calcula a contagem para a loja atual
    contagem_bruta = contar_categorias(lista_vendas)

    # 2. Ordena a contagem
    contagem_ordenada = ordenar_contagem(contagem_bruta)

    # 3. Armazena o resultado final
    resultados_categorias_ordenados[nome_loja] = contagem_ordenada

    # 4. Imprime de forma formatada
    print(f"\n{nome_loja.upper()}:")
    print(contagem_ordenada)

# 'resultados_categorias_ordenados' agora contém todos os dados
# prontos para gráficos de pizza ou barras.
```

```python
import matplotlib.pyplot as plt

# --- 1. Função Helper (mantida igual) ---
def preparar_dados_barra(lista_ordenada: list, top_n: int = 5):
    """
    Prepara dados para gráfico de barras horizontais,
    pegando o Top N e agrupando o restante em 'Outros'.
    """
    top_items = lista_ordenada[:top_n]
    outros_items = lista_ordenada[top_n:]

    categorias = [item[0] for item in top_items]
    contagens = [item[1] for item in top_items]

    if outros_items:
        contagem_outros = sum(item[1] for item in outros_items)
        if contagem_outros > 0:
            categorias.append(f'Outros ({len(outros_items)} categ.)')
            contagens.append(contagem_outros)

    return categorias, contagens


# --- 2. Selecionar loja e preparar dados ---
loja_alvo = 'Loja 4'  # 🏪 Altere aqui para a loja desejada
n_fatias = 5           # Top 5 categorias principais

dados_loja = resultados_categorias_ordenados[loja_alvo]
categorias, contagens = preparar_dados_barra(dados_loja, top_n=n_fatias)

# --- 3. Criar gráfico de barras horizontais ---
plt.figure(figsize=(10, 6))
cores = plt.cm.Blues_r(range(len(categorias)))  # Gradiente de azul

plt.barh(categorias, contagens, color=cores)
plt.gca().invert_yaxis()  # Inverte a ordem (maiores no topo)

# --- 4. Adicionar rótulos e título ---
plt.title(f'Vendas por Categoria - {loja_alvo}', fontsize=16)
plt.xlabel('Quantidade de Vendas', fontsize=12)
plt.ylabel('Categoria do Produto', fontsize=12)

# --- 5. Mostrar valores à direita das barras ---
for i, valor in enumerate(contagens):
    plt.text(valor + 0.3, i, str(valor), va='center', fontsize=11)

plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
# 1. Função: gerar ranking de produtos
def gerar_ranking_produtos(vendas: list) -> list:
    """
    Gera um ranking de produtos com base na quantidade de vendas.
    Retorna uma lista de tuplas (produto, quantidade), ordenada do mais vendido ao menos vendido.
    """
    contagem = {}

    # Conta quantas vezes cada produto aparece nas vendas
    for item in vendas:
        nome_produto = item['Produto']
        contagem[nome_produto] = contagem.get(nome_produto, 0) + 1

    # Ordena pelo número de vendas (do maior para o menor)
    ranking = sorted(contagem.items(), key=lambda x: x[1], reverse=True)
    return ranking
```

```python
# 1. Função para gerar o ranking de produtos ----------------------

def gerar_ranking_produtos(lista_vendas: list) -> list:
    """
    Gera um ranking dos produtos mais e menos vendidos de uma loja.
    Retorna uma lista ordenada (produto, quantidade).
    """
    contagem = {}

    # Conta quantas vezes cada produto aparece nas vendas
    for venda in lista_vendas:
        produto = venda['Produto']
        contagem[produto] = contagem.get(produto, 0) + 1

    # Ordena do mais vendido para o menos vendido
    ranking = sorted(contagem.items(), key=lambda item: item[1], reverse=True)
    return ranking


# 2. Cálculo do ranking para todas as lojas -----------------------

print("\n=== Ranking de Produtos: Mais e Menos Vendidos ===")

ranking_por_loja = {}

for nome_loja, lista_vendas in vendas_lojas.items():
    # Gera o ranking da loja atual
    ranking = gerar_ranking_produtos(lista_vendas)
    ranking_por_loja[nome_loja] = ranking

    # Exibição dos resultados
    print(f"\n📍 {nome_loja}")
    print("  🏆 Top 5 mais vendidos:")
    for produto, qtd in ranking[:5]:
        print(f"    - {produto}: {qtd} vendas")

    print("  ⚠️ 5 menos vendidos:")
    for produto, qtd in ranking[-5:]:
        print(f"    - {produto}: {qtd} vendas")


# 3. Gráfico com o Top 5 da Loja 1 ----------

import matplotlib.pyplot as plt

loja1_top5 = ranking_por_loja["Loja 1"][:5]
produtos, quantidades = zip(*loja1_top5)

plt.figure(figsize=(8, 5))
plt.barh(produtos, quantidades, color='royalblue')
plt.xlabel("Quantidade Vendida")
plt.title("Top 5 Produtos Mais Vendidos - Loja 1")
plt.gca().invert_yaxis()  # Inverte para mostrar o mais vendido no topo
plt.show()
```

# 5. Frete Médio por Loja

```python
# 1. Função para calcular a média de frete -------------------------
def calcular_media_frete(lista_vendas: list) -> float:
    """
    Calcula o valor médio do frete por venda em uma loja.
    """
    soma_frete = 0.0

    for venda in lista_vendas:
        soma_frete += venda['Frete']

    total_vendas = len(lista_vendas)
    return soma_frete / total_vendas if total_vendas > 0 else 0.0


# 2. Cálculo e exibição dos resultados -----------------------------
print("=== Frete Médio por Venda (R$) ===")

media_frete_por_loja = {}

for nome_loja, lista_vendas in vendas_lojas.items():
    media = calcular_media_frete(lista_vendas)
    media_frete_por_loja[nome_loja] = media
    print(f"{nome_loja}: R$ {media:.2f}")


# 3. Gráfico comparativo entre as lojas -----------------------------
import matplotlib.pyplot as plt

# Extrair dados
lojas = list(media_frete_por_loja.keys())
medias = list(media_frete_por_loja.values())

# Criar gráfico de colunas
plt.figure(figsize=(8, 5))
barras = plt.bar(lojas, medias, color=['#E74C3C', '#F39C12', '#3498DB', '#2ECC71'])

# Adicionar rótulos de valor no topo das barras
for barra in barras:
    plt.text(
        barra.get_x() + barra.get_width() / 2,
        barra.get_height(),
        f"R$ {barra.get_height():.2f}",
        ha='center', va='bottom', fontsize=10, fontweight='bold'
    )

# Personalização visual
plt.title("Frete Médio por Loja", fontsize=16)
plt.ylabel("Valor Médio (R$)")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
# ===============================================
# RELATÓRIO FINAL DE ANÁLISE: LOJAS ALURA STORE
# ===============================================

import pandas as pd
import matplotlib.pyplot as plt

# --- 1. Simulação dos DataFrames (coloque aqui os seus dados reais) ---
# Se você já tem loja1, loja2, loja3, loja4 carregados, pule esta célula.
# Exemplo:
# loja1 = pd.read_csv("loja1.csv")

# --- 2. Agrupar as lojas em um dicionário ---
lojas_df = {
    "Loja 1": loja1,
    "Loja 2": loja2,
    "Loja 3": loja3,
    "Loja 4": loja4
}

# Converter DataFrames em listas de dicionários (para os cálculos)
vendas_lojas = {nome: df.to_dict(orient="records") for nome, df in lojas_df.items()}


# =====================================================
# 3. FATURAMENTO TOTAL
# =====================================================
def obter_faturamento(vendas: list) -> float:
    return sum(v["Preço"] for v in vendas)

faturamentos = {loja: obter_faturamento(vendas) for loja, vendas in vendas_lojas.items()}

print("=== Faturamento Total por Loja ===")
for loja, valor in faturamentos.items():
    print(f"{loja}: R$ {valor:,.2f}")

# --- Gráfico de barras ---
plt.figure(figsize=(8, 5))
plt.bar(faturamentos.keys(), faturamentos.values(), color=["#E74C3C", "#F39C12", "#3498DB", "#2ECC71"])
plt.title("Faturamento Total por Loja", fontsize=16)
plt.ylabel("Faturamento (R$)")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()


# =====================================================
# 4. VENDAS POR CATEGORIA
# =====================================================
def contar_categorias(lista_vendas: list) -> dict:
    contagem = {}
    for venda in lista_vendas:
        cat = venda["Categoria do Produto"].lower()
        contagem[cat] = contagem.get(cat, 0) + 1
    return contagem

resultados_categorias = {loja: contar_categorias(vendas) for loja, vendas in vendas_lojas.items()}

print("\n--- Contagem de Vendas por Categoria ---")
for loja, categorias in resultados_categorias.items():
    print(f"\n{loja.upper()}:\n{categorias}")

# --- Ordenar categorias ---
def ordenar_contagem(d: dict):
    return sorted(d.items(), key=lambda x: x[1], reverse=True)

resultados_categorias_ordenados = {loja: ordenar_contagem(cat) for loja, cat in resultados_categorias.items()}

print("\n--- Vendas por Categoria (Mais Populares) ---")
for loja, ranking in resultados_categorias_ordenados.items():
    print(f"\n{loja.upper()}:\n{ranking}")


# =====================================================
# 5. MÉDIA DE AVALIAÇÃO DAS LOJAS
# =====================================================
def calcular_media_avaliacao(vendas: list) -> float:
    if not vendas:
        return 0
    soma = sum(v["Avaliação da compra"] for v in vendas)
    return soma / len(vendas)

resultados_media_avaliacao = {loja: calcular_media_avaliacao(vendas) for loja, vendas in vendas_lojas.items()}

print("\n=== Média de Avaliação das Lojas ===")
for loja, media in resultados_media_avaliacao.items():
    print(f"{loja}: {media:.2f} estrelas")

# --- Gráfico de barras ---
plt.figure(figsize=(8,5))
plt.bar(resultados_media_avaliacao.keys(), resultados_media_avaliacao.values(), color=["#E74C3C", "#F39C12", "#3498DB", "#2ECC71"])
plt.title("Média de Avaliação por Loja", fontsize=16)
plt.ylabel("Avaliação Média (1–5 estrelas)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()


# =====================================================
# 6. RANKING DE PRODUTOS MAIS E MENOS VENDIDOS
# =====================================================
def rankear_produtos(vendas: list) -> list:
    contagem = {}
    for v in vendas:
        produto = v["Produto"]
        contagem[produto] = contagem.get(produto, 0) + 1
    return sorted(contagem.items(), key=lambda x: x[1], reverse=True)

ranking_produtos = {loja: rankear_produtos(vendas) for loja, vendas in vendas_lojas.items()}

print("\n=== Ranking de Produtos: Mais e Menos Vendidos ===")
for loja, ranking in ranking_produtos.items():
    print(f"\n{loja.upper()}:")
    print(f"  Mais vendidos: {ranking[:5]}")
    print(f"  Menos vendidos: {ranking[-5:]}")


# =====================================================
# 7. FRETE MÉDIO POR LOJA
# =====================================================
def calcular_media_frete(vendas: list) -> float:
    if not vendas:
        return 0
    soma = sum(v["Frete"] for v in vendas)
    return soma / len(vendas)

media_frete_por_loja = {loja: calcular_media_frete(vendas) for loja, vendas in vendas_lojas.items()}

print("\n=== Frete Médio por Loja ===")
for loja, valor in media_frete_por_loja.items():
    print(f"{loja}: R$ {valor:.2f}")

# --- Gráfico de colunas ---
plt.figure(figsize=(8,5))
barras = plt.bar(media_frete_por_loja.keys(), media_frete_por_loja.values(), color=["#E74C3C", "#F39C12", "#3498DB", "#2ECC71"])
for barra in barras:
    plt.text(barra.get_x() + barra.get_width()/2, barra.get_height(), f"R$ {barra.get_height():.2f}", ha='center', va='bottom', fontsize=10)
plt.title("Frete Médio por Loja", fontsize=16)
plt.ylabel("Valor Médio (R$)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()


# =====================================================
# 8. SUMÁRIO FINAL
# =====================================================
print("\n========== RELATÓRIO FINAL ==========")
print("🔹 Faturamento total por loja:", faturamentos)
print("🔹 Média de avaliação por loja:", resultados_media_avaliacao)
print("🔹 Frete médio por loja:", media_frete_por_loja)
print("🔹 Categorias mais vendidas por loja:", {k: v[:3] for k,v in resultados_categorias_ordenados.items()})
print("🔹 Top produtos por loja:", {k: v[:3] for k,v in ranking_produtos.items()})
print("=====================================")
```
